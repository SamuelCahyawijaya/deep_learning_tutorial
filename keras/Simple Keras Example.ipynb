# Simple Keras Example
Please go to https://keras.io/ for the package documentation

## Load Common Package

```python
import numpy as np
```

## Load Keras Package

```python
import keras
from keras.models import Model, Sequential # We can use both of this to build a model
from keras.layers import Input, Dense, Dropout, Activation, BatchNormalization
from keras.optimizers import SGD, Adam
```

## Basic Fully Connected Layer With Single Output
<img src="https://cdn-images-1.medium.com/max/1600/1*BMdnuQUqGEb_FSEmYIxw5g.jpeg" width="50%"/>

##### Generate Dummy Data

```python
np.random.seed(0)

x_train = np.random.random((500, 3))
y_train = np.mean(x_train, axis=-1)
x_test = np.random.random((50, 3))
y_test = np.mean(x_test, axis=-1)
```

```python
print("x_train.shape",x_train.shape)
print("y_train.shape",y_train.shape)
print("x_test.shape",x_test.shape)
print("y_test.shape",y_test.shape)
print("x_train[0]",x_train[0])
print("y_train[0]",y_train[0])
print("x_test[0]",x_test[0])
print("y_test[0]",y_test[0])
```

##### Initialize Model

```python
sequential_model = Sequential()
```

```python
sequential_model.add(Dense(units=4, activation='relu', input_dim=3))
sequential_model.add(Dense(units=1, input_dim=4))
sequential_model.add(Activation('tanh'))
sequential_model.summary()
```

```python
sequential_model.compile(loss='mae', optimizer='adam', metrics=['mse'])
```

##### Train Model

```python
sequential_model.fit(x_train, y_train, epochs=200, batch_size=50)
```

##### Evaluate on Test Data

```python
score = sequential_model.evaluate(x_test, y_test)
print(score)
```

```python
predict = sequential_model.predict(x_test)
for i in range(0, len(x_test)):
    print(x_test[i], y_test[i], predict[i])
```

## Let's try with Deeper Network
<img src="http://cs231n.github.io/assets/nn1/neural_net2.jpeg" width="50%"/>

##### Initialize Model

```python
inputs = Input(shape=(3,))
x = Dense(units=7, activation='relu', input_dim=3)(inputs)
x = Dense(units=7, activation='relu', input_dim=7)(x)
x = Dense(units=7, activation='relu', input_dim=7)(x)
x = Dense(units=7, activation='relu', input_dim=7)(x)
x = Dense(units=7, activation='relu', input_dim=7)(x)
x = Dense(units=1, input_dim=7)(x)
outputs = Activation('tanh')(x)
```

```python
model.compile(loss='mae', optimizer='adam', metrics=['mse'])
```

##### Train Model

```python
model.fit(x_train, y_train, epochs=200, batch_size=50)
```

##### Evaluate on Test Data

```python
score = model.evaluate(x_test, y_test)
print(score)
```

```python
predict = model.predict(x_test)
for i in range(0, len(x_test)):
    print(x_test[i], y_test[i], predict[i])
```

## Neural Network With Softmax Output
<img src="http://paddlepaddle.org/docs/develop/book/02.recognize_digits/image/softmax_regression_en.png" width="30%"/>

##### Generate Dummy Data

```python
def generate_categorical_y(x):
    return ((x[:,0] > 0.5) * 4) + ((x[:,1] > 0.5) * 2) + (x[:,2] > 0.5)
```

```python
np.random.seed(0)

x_train = np.random.random((1000, 3))
y_train = generate_categorical_y(x_train)
print(x_train[0], y_train[0])
print(x_train[7], y_train[7])
print(x_train[83], y_train[83])
print(x_train[83], y_train[299])
```

```python
np.random.seed(0)

x_train = np.random.random((500, 3))
y_train = keras.utils.to_categorical(generate_categorical_y(x_train), num_classes=8)
x_test = np.random.random((50, 3))
y_test = keras.utils.to_categorical(generate_categorical_y(x_test), num_classes=8)
```

```python
print("x_train.shape",x_train.shape)
print("y_train.shape",y_train.shape)
print("x_test.shape",x_test.shape)
print("y_test.shape",y_test.shape)
print("x_train[0]",x_train[0])
print("y_train[0]",y_train[0])
print("x_test[0]",x_train[83])
print("y_test[0]",y_train[83])
```

##### Initialize Model

```python
inputs = Input(shape=(3,))
x = Dense(units=7, activation='relu', input_dim=3)(inputs)
x = Dense(units=7, activation='relu', input_dim=7)(x)
x = Dense(units=7, activation='relu', input_dim=7)(x)
x = Dense(units=7, activation='relu', input_dim=7)(x)
x = Dense(units=7, activation='relu', input_dim=7)(x)
x = Dense(units=8, input_dim=7)(x)
outputs = Activation('softmax')(x)
```

```python
model = Model(inputs=inputs, outputs=outputs)
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

##### Train Model

```python
model.fit(x_train, y_train, epochs=500, batch_size=50)
```

##### Evaluate on Test Data

```python
score = model.evaluate(x_test, y_test)
print(score)
```

```python
predict = model.predict(x_test)
for i in range(0, len(y_test)):
    print(y_test[i], predict[i])
```

## Let's try adding Batch Normalization
<img src="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQnlFLeKkzpQLzxCQSKeNYc6SVtzQIrwGDLD-v214WJQNbu0cfi" width="50%"/>

##### Initialize Model

```python
inputs = Input(shape=(3,))
x = Dense(units=7, input_dim=3)(inputs)
x = BatchNormalization()(x)
x = Activation('relu')(x)
x = Dense(units=7, input_dim=7)(x)
x = BatchNormalization()(x)
x = Activation('relu')(x)
x = Dense(units=7, input_dim=7)(x)
x = BatchNormalization()(x)
x = Activation('relu')(x)
x = Dropout(0.1)(x)
x = Dense(units=7, input_dim=7)(x)
x = BatchNormalization()(x)
x = Activation('relu')(x)
x = Dense(units=7, input_dim=7)(x)
x = BatchNormalization()(x)
x = Activation('relu')(x)
x = Dense(units=8, input_dim=7)(x)
x = BatchNormalization()(x)
outputs = Activation('softmax')(x)
```

```python
model = Model(inputs=inputs, outputs=outputs)
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

##### Train Model

```python
model.fit(x_train, y_train, epochs=500, batch_size=50)
```

##### Evaluate on Test Data

```python
score = model.evaluate(x_test, y_test)
print(score)
```

```python
predict = model.predict(x_test)
for i in range(0, len(y_test)):
    print(y_test[i], predict[i])
```
